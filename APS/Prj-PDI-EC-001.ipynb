```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import skimage as ski
from skimage import io
from google.colab import drive
import math
from google.colab import files

def read_image(url):
  return ski.io.imread(url)

def equalize_image(url):
  # imagem, imagem preto e branco e imagem equalizada
  image = read_image(url)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  equ = cv2.equalizeHist(gray) 

  # calculo da variancia
  imean = np.mean(gray)
  grayf = gray.astype(np.float32)
  grayf2 = grayf * grayf
  imeanf2 = np.mean(grayf2)
  ivar = imeanf2 - imean**2
  istd = math.sqrt(ivar)

  # plota a imagem equalizada
  fig, ax = plt.subplots()
  plt.title("Imagem equalizada - Variance " + str(ivar))
  plt.imshow(equ, cmap='gray')

  index = url.rfind('/')
  new_name = url[index+1:-4] + '-' + 'equ-var' + '.png'
  plt.savefig(new_name)
  files.download(new_name) 

def histogram_image(url):
  # imagem
  image = read_image(url)

  # gera e plota o histograma
  colors = ("red", "green", "blue")
  plt.figure()
  plt.xlim([0, 256])
  for channel_id, color in enumerate(colors):
      histogram, bin_edges = np.histogram(
          image[:, :, channel_id], bins=256, range=(0, 256)
      )
      plt.plot(bin_edges[0:-1], histogram, color=color)

  plt.title("Histograma")
  plt.xlabel("Valor da cor")
  plt.ylabel("Número de pixels")

  index = url.rfind('/')
  new_name = url[index+1:-4] + '-' + 'hist' + '.png'
  plt.savefig(new_name)
  files.download(new_name)

fruits = [('pineapple', '08')]
versions = ['V1', 'V2']
colors = ['B', 'W']
operations = ['', 'cnv', 'pot', 'mirror']
url_original_images = []
url_extended_images = []

drive.mount('/content/drive')

# Save the images
url_base = 'https://raw.githubusercontent.com/dev-marcos/Digital-image-processing/main/ft-DB/720p/'
for fruit in fruits:
  url_fruta = url_base + fruit[0] + '/0-' + fruit[1] + '-'
  for version in versions:
    url_version = url_fruta + version + '-'
    for color in colors:
      url_color = url_version + color + '.png'
      url_original_images.append(url_color)

url_base = 'https://raw.githubusercontent.com/Fellipesre/pdi/main/'
for fruit in fruits:
  url_fruta = url_base + fruit[0] + '/0-' + fruit[1] + '-'
  for version in versions:
    url_version = url_fruta + version + '-'
    for color in colors:
      url_color = url_version + color
      for operation in operations:
        if operation != '':
          url_operation = url_color + '-' + operation + '.png'
        else:
          url_operation = url_color + '.png'
        url_extended_images.append(url_operation)

'''
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V1-W-log.png', cv2.cvtColor(log_image1, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V1-W-pot.png', cv2.cvtColor(pot_image1, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V1-W-cnv.png', cv2.cvtColor(cnv_image1, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V1-B-log.png', cv2.cvtColor(log_image2, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V1-B-pot.png', cv2.cvtColor(pot_image2, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V1-B-cnv.png', cv2.cvtColor(cnv_image2, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V2-W-log.png', cv2.cvtColor(log_image3, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V2-W-pot.png', cv2.cvtColor(pot_image3, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V2-W-cnv.png', cv2.cvtColor(cnv_image3, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V2-B-log.png', cv2.cvtColor(log_image4, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V2-B-pot.png', cv2.cvtColor(pot_image4, cv2.COLOR_RGB2BGR))
cv2.imwrite('/content/drive/MyDrive/PDI/0-08-V2-B-cnv.png', cv2.cvtColor(cnv_image4, cv2.COLOR_RGB2BGR))
'''

for url in url_extended_images:
  equalize_image(url)
  #histogram_image(url)

  
```
