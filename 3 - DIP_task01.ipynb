Antes de mais nada vamos importar as bibliotecas ``numpy``, ``skimage`` e ``matplotlib`` (biblioteca para visualiza√ß√£o). Vamos tamb√©m criar a fun√ß√£o ``show_image`` para visualiza√ß√£o das imagens na maioria dos casos. Essa fun√ß√£o √© baseada na fun√ß√£o nativa ``matplotlib.pyplot.imshow``.

```python
import numpy as np
import skimage as ski

import matplotlib.pyplot as plt

# versions
print(f'numpy:{np.__version__}\nskimage:{ski.__version__}')

def show_image(image, title='Image', cmap_type='gray', axis='off'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis(axis)
```

1) **Importe a imagem *gravel* (cascalho). Exiba as dimens√µes da imagem original. Crie uma nova imagem sobrepondo um reticulado preto de $10 \times 10$ pixels. Apresente a imagem original e a nova imagem lado a lado.**

üí° **Dica: n√£o esque√ßa de copiar a imagem original antes aplicar as opera√ß√µes.**

```python
img_gravel = ski.data.gravel()

print(f"Original image dimensions: {img_gravel.shape[0]} x {img_gravel.shape[1]} pixels")

#  insira sua resposta

img_r = img_gravel.copy()  # array copy

img_r[::10] = 0
img_r[::,::10] = 0


fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img_gravel, cmap=plt.cm.gray)
ax[0].set_title("Original")
ax[1].imshow(img_r, cmap=plt.cm.gray)
ax[1].set_title("Image with reticle")

fig.tight_layout()
plt.show()
```

3) **Importe a imagem *camera*. Crie um recorte com todas as linhas e as colunas de $128$ a $360$. Exiba as dimens√µes da imagem original e do recorte. Crie uma nova imagem sendo a transposta da imagem original. Exiba as dimens√µes da imagem transposta. Apresente a imagem original e a nova imagem lado a lado.**


```python
img_camera = ski.data.camera()

print(f"Original image dimensions: {img_camera.shape[0]} x {img_camera.shape[1]} pixels")

img_camera = img_camera[:,128:360]

print(f"Clipped image dimensions: {img_camera.shape[0]} x {img_camera.shape[1]} pixels")


# Flip the previous image horizontally
img_camera_horizontal_flip = np.fliplr(img_camera)

# rotate
img_camera_horizontal_flip_rotated = np.rot90(img_camera_horizontal_flip, axes=(0, 1))

print(f"Tranpose image dimensions: {img_camera_horizontal_flip_rotated.shape[0]} x {img_camera_horizontal_flip_rotated.shape[1]} pixels")


fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img_camera, cmap=plt.cm.gray)
ax[0].set_title("Original")
ax[1].imshow(img_camera_horizontal_flip_rotated, cmap=plt.cm.gray)
ax[1].set_title("Transposed")

fig.tight_layout()
plt.show()

```

Crie uma nova imagem fazendo uma rota√ß√£o antihor√°ria de $90^o$ da imagem original. Apresente as tr√™s imagens (recorte, transposta e rotacionada) lado a lado.**

üí° **Dica: uma das t√©cnicas mais simples para realizar a rota√ß√£o antihor√°ria de uma matriz √© calcular a transposta e depois refleti-la na vertical.**


```python
img_camera = ski.data.camera()

print(f"Original image dimensions: {img_camera.shape[0]} x {img_camera.shape[1]} pixels")


#  insira sua resposta
img_camera = img_camera[:,128:360]

print(f"Clipped image dimensions: {img_camera.shape[0]} x {img_camera.shape[1]} pixels")



# rotate
img_camera_rotated = np.rot90(img_camera, axes=(0, 1))

# Flip the previous image horizontally
img_camera_rotated_horizontal_flip = np.flipud(img_camera_rotated)



print(f"Tranpose image dimensions: {img_camera_rotated_horizontal_flip.shape[0]} x {img_camera_rotated_horizontal_flip.shape[1]} pixels")


fig, axes = plt.subplots(1, 3, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img_camera, cmap=plt.cm.gray)
ax[0].set_title("Original")
ax[1].imshow(img_camera_rotated_horizontal_flip, cmap=plt.cm.gray)
ax[1].set_title("Transposed")
ax[2].imshow(img_camera_rotated, cmap=plt.cm.gray)
ax[2].set_title("Original Rotated")

fig.tight_layout()
plt.show()
```


5) **Importe a imagem *camera*. Crie uma subamostragem $1/4$ (*subsampling*) utilizando *slicing*. Exiba as dimens√µes da imagem original e da subamostragem.  Apresente as imagens original e subamostrada lado a lado.**


```python
img_camera = ski.data.camera()

print(f"Original image dimensions: {img_camera.shape[0]} x {img_camera.shape[1]} pixels")

#  insira sua resposta

from skimage import transform

img_camera_resized = transform.resize(img_camera, (img_camera.shape[0]/4, img_camera.shape[1]/4), anti_aliasing=False)

print(f"Subsample dimensions: {img_camera_resized.shape[0]} x {img_camera_resized.shape[1]} pixels")


fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img_camera, cmap=plt.cm.gray)
ax[0].set_title("Original")
ax[1].imshow(img_camera_resized, cmap=plt.cm.gray)
ax[1].set_title("Subsample")

fig.tight_layout()
plt.show()



```

6) **Importe a imagem *horse*. Crie uma imagem ampliada $2\times$ por replica√ß√£o, i.e, copiando os pixels vizinhos.  Apresente as imagens original e a ampliada lado a lado.**


```python
img_horse = ski.data.horse()

print(f"Original image dimensions: {img_horse.shape[0]} x {img_horse.shape[1]} pixels")

#  insira sua resposta

from skimage import transform

img_horse_resized = transform.resize(img_horse, (img_horse.shape[0]*2, img_horse.shape[1]*2), anti_aliasing=False)

print(f"Enlarged dimensions: {img_horse_resized.shape[0]} x {img_horse_resized.shape[1]} pixels")


fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img_horse, cmap=plt.cm.gray)
ax[0].set_title("Original")
ax[1].imshow(img_horse_resized, cmap=plt.cm.gray)
ax[1].set_title("Enlarged 2x")

fig.tight_layout()
plt.show()
```

7) **Importe a imagem *logo*. Converta para *grayscale*. Crie uma imagem duplicando as dimens√µes da imagem original. No canto superior esquerdo insira a imagem original. No canto superior direito insira a reflex√£o vertical. No canto inferior esquerdo a reflex√£o na vertical. No canto inferior direito a reflex√£o horizontal e vertical.**


```python
# Convert the image to grayscale
img_logo = ski.color.rgb2gray(ski.color.rgba2rgb(ski.data.logo()))

print(f"Original image dimensions: {img_logo.shape[0]} x {img_logo.shape[1]} pixels")

#  insira sua resposta
img_comp = np.empty((img_logo.shape[0]*2, img_logo.shape[1]*2))

img_comp[:img_logo.shape[0],:img_logo.shape[1]] = img_logo
img_comp[:img_logo.shape[0],img_logo.shape[1]:] = np.fliplr(img_logo)
img_comp[img_logo.shape[0]:,:img_logo.shape[1]] = np.flipud(img_logo)
img_comp[img_logo.shape[0]:,img_logo.shape[1]:] = np.flip(img_logo)

print(f"Composition dimensions: {img_comp.shape[0]} x {img_comp.shape[1]} pixels")

show_image(img_comp, "Reflections", axis='on')
```

8) **Transla√ß√£o √© uma opera√ß√£o primitiva utilizada para implementar outras opera√ß√µes. Nessa opera√ß√£o os pixels s√£o deslocados por um fator na vertical e na horizontal. Importe a imagem *horse*. Crie uma imagem com $2.5 \times$ as dimens√µes dessa imagem. Sobreponha um reticulado preto de $10 \times 10$ pixels. Insira a imagem *horse* em $0,0$. Fa√ßa uma transla√ß√£o de 330 linhas e 200 colunas.**



```python
img_horse = ski.data.horse().astype(np.int64)

print(f"Original image dimensions: {img_horse.shape[0]} x {img_horse.shape[1]} pixels")

#  insira sua resposta
img_comp = np.full((int(img_horse.shape[0]*2.5), int(img_horse.shape[1]*2.5)),1)
#img_comp = np.empty((int(img_horse.shape[0]*2.5), int(img_horse.shape[1]*2.5)))


img_comp[::10] = 0
img_comp[::,::10] = 0


img_comp[:img_horse.shape[0],:img_horse.shape[1]] = img_horse.copy()

img_comp[330:330+img_horse.shape[0],200:200+img_horse.shape[1]] = img_comp[:img_horse.shape[0],:img_horse.shape[1]]


print(f"Composition dimensions: {img_comp.shape[0]} x {img_comp.shape[1]} pixels")

show_image(img_comp, "Translation", axis='on')
```

9)** Importe as images *camera* e *brick*. Crie uma composi√ß√£o combinando a primeira imagem com os pares da segunda.**



```python
img_camera = ski.data.camera()
img_brick = ski.data.brick()

print(f"Original image camera dimensions: {img_camera.shape[0]} x {img_camera.shape[1]} pixels")
print(f"Original image brick dimensions: {img_brick.shape[0]} x {img_brick.shape[1]} pixels")


#  insira sua resposta
img_s = img_camera.copy()
img_s[::2] = img_brick[::2]
print(f"Composition dimensions: {img_s.shape[0]} x {img_s.shape[1]} pixels")

show_image(img_s, "Composition", axis='on')
```

10) **Utilizando a imagem gerada ao final do exerc√≠cio anterior, separe os campos pares e √≠mpares. Apresente as tr√™s imagens.**

```python
img_even = np.zeros_like(img_s)

print(f"Original image dimensions: {img_s.shape[0]} x {img_s.shape[1]} pixels")

#  insira sua resposta

img_even[::2] = img_s[::2]
img_even[1::2] = img_s[::2]

img_odd = np.zeros_like(img_s)
img_odd[::2] = img_s[1::2]
img_odd[1::2] = img_s[1::2]

fig, axes = plt.subplots(1, 3, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(img_s, cmap=plt.cm.gray)
ax[0].set_title("Original interleaved")
ax[1].imshow(img_even, cmap=plt.cm.gray)
ax[1].set_title("Even")
ax[2].imshow(img_odd, cmap=plt.cm.gray)
ax[2].set_title("Odd")

fig.tight_layout()
plt.show()
```

11) **Uma imagem rampa (*ramp*) caracteriza-se por apresentar uma varia√ß√£o crescente de n√≠veis de cinza, i.e., dos mais baixos (escuros) para os mais altos (claros). Normalmente  √© utilizada para teste de algoritmos.**

**Numericamente podemos gerar uma rampa unidimensional (1-D) com ``np.arange``**



```python
n = 10
ramp = np.arange(n)

print(f'ramp 1-D > {ramp.shape} \n{ramp}')
```

**Uma rampa bidimensional (2-D) pode ser criada atrav√©s da fun√ß√£o ``np.resize``, que altera o *array* para o novo *shape*. Se a quantidade de elementos for maior o *array* √© complementado  atrav√©s da replica√ß√£o dos elementos que s√£o concatenados linearmente.**

```python
ramp2 = np.resize(ramp, (n, n))

print(f'ramp 2-D > {ramp2.shape} \n{ramp2}')
```

**Baseando-se no c√≥digo anterior, crie e apresente imagens rampa com $n = {2, 4, 8, 16, 32, 64, 128, 256}$.**

```python
#  insira sua resposta

fig, axes = plt.subplots(2, 4, figsize=(8, 4))
ax = axes.ravel()

for x in range(8):
  n = 2**(x+1)

  ax[x].imshow(np.resize(np.arange(n), (n, n)), cmap=plt.cm.gray)
  ax[x].set_title(f"Ramp {n}")

fig.tight_layout()
plt.show()
```
